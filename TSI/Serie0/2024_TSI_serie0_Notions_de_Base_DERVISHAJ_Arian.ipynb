# Traitement du signal et d'images
# Série 0 : La dimensionnalité des signaux

La dimensionnalité du signal correspond **au nombre de ses dépendances**.<br>
En d'autres termes, la dimensionnalité correspond au nombre de variables dont dépend la fonction mathématique qui modélise le signal.<br>
Attention: les variables ne sont pas les paramètres de la fonction mathématique.<br>
<br>
<br>
Exemples:<br>
$u(t)$ => Signal 1D<br>
$u(x,y)$ => Signal 2D<br>
$u(x,y,z)$ ou $u(x,y,t)$ => Signal 3D<br>
$u(x,y,z,t)$ => Signal 4D<br>
$u(x_1,x_2,...,x_N)$ => Signal ND<br>

Soit le signal $x(t)$ ci-dessous.

```python
# Paramètres
fe=8000 # Hz (fréquence d'échantillonnage)
f0= 440 # Hz (fréquence du signal)

# Création des vecteurs temps et amplitude
Te=1/fe # s (période d'échantillonnage)
T=1/f0 # s (période du signal)
tmin=0 # s (temps initial)
tmax=3*T # s (temps final = 3 périodes)
t=np.arange(tmin, tmax, Te)# vecteur temps
x=np.sin(2*np.pi*f0*t)

# Affichage du signal
plt.stem(t, x) # rajouter use_line_collection avec la dernière version matplotlib
plt.xlabel('temps (s)')
plt.title('x(t)')
plt.grid(True)
plt.show()
```
